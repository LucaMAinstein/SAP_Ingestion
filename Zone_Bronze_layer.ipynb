```python
from pyspark.sql.functions import current_date, current_timestamp, expr

from pyspark.sql.functions import col

```

Creating Bronze Layer: Zone 1

```python
#Load data

zone = "Zone1_raw"

df = spark.read.table(zone).withColumn("load_time_GMT", current_timestamp())

df = df.drop("Project1")

rename_dict = {
    "Code1": "WBS",
    "SAP Internal Number1": "sap_internal_number",
    "SAP Source1": "SAP_source",
    "Currency1": "Currency",
    "Year1": "Year",
    "Budget _LC_1": "BGT_LC",
    "Actual _LC_1": "ACT_LC",
    "Commitment _LC_1": "COM_LC"
}

ordem = [
    "Source_name","WBS", "sap_internal_number", "SAP_source", 
    "Currency", "Year", "BGT_LC", "ACT_LC", "COM_LC", "load_time_GMT"
]


df_renom = df.select([
    col(c).alias(rename_dict.get(c, c))
    for c in df.columns
])

df_renom = df_renom.select(*ordem)

# Definindo o caminho do destino (modifique conforme o seu caso)
caminho = "abfss://SAP_Ingest_v2@onelake.dfs.fabric.microsoft.com/Bronze.Lakehouse/Tables/dbo/Bronze/Zone1/bronze_zone1"

# Salvando o DataFrame no formato Delta
df_renom.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .save(caminho)

print("Dados salvos com sucesso!")

# Executando o comando OPTIMIZE para otimizar a tabela Delta
spark.sql(f"OPTIMIZE delta.`{caminho}`")

# Confirmação de que a otimização foi concluída
print(f"Optimization for {caminho} completed!")
```
