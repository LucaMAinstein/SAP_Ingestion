```python
from pyspark.sql.functions import current_date, current_timestamp, expr

from pyspark.sql.functions import col
```

```python
#Load data

zone = "IO_Project"

df = spark.read.table(zone).withColumn("load_time_GMT", current_timestamp())

# Definindo o caminho do destino (modifique conforme o seu caso)
caminho = "abfss://SAP_Ingest_v2@onelake.dfs.fabric.microsoft.com/Bronze.Lakehouse/Tables/dbo/Bronze/IO_Project/bronze_IO_Project"

rename_dict = {
    "IO": "WBS",
    "SAP Internal Number": "sap_internal_number",
    "Project": "Project_BA"
}


df_renom = df.select([
    col(c).alias(rename_dict.get(c, c))
    for c in df.columns
])

# Salvando o DataFrame no formato Delta
df_renom.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .save(caminho)

print("Dados salvos com sucesso!")

# Executando o comando OPTIMIZE para otimizar a tabela Delta
spark.sql(f"OPTIMIZE delta.`{caminho}`")

# Confirmação de que a otimização foi concluída
print(f"Optimization for {caminho} completed!")
```
